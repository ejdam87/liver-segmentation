```python
import numpy as np
import cv2
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Training U-Net

metrics = pd.read_csv('/content/metric_values_unet.csv')
epochs = np.array([i for i in range(1, metrics.shape[0]+1)])

plt.figure()
plt.plot(epochs, metrics["mPA"], label="mPA")
plt.plot(epochs, metrics["mIoU"], label="mIoU")
plt.xlabel("Epochs")
plt.ylabel("Metrics")
plt.title("mPA and mIoU during U-Net training")
plt.legend()
plt.show()

plt.figure()
plt.plot(epochs, metrics["train_loss"], label="training loss")
plt.plot(epochs, metrics["val_loss"], label="validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("U-Net training and validation loss")
plt.legend()
plt.show()
```

```python
# Training baseline

metrics = pd.read_csv('/content/metric_values_baseline.csv')
epochs = np.array([i for i in range(1, metrics.shape[0]+1)])

plt.figure()
plt.plot(epochs, metrics["mPA"], label="mPA")
plt.plot(epochs, metrics["mIoU"], label="mIoU")
plt.xlabel("Epochs")
plt.ylabel("Metrics")
plt.title("mPA and mIoU during baseline training")
plt.legend()
plt.show()

plt.figure()
plt.plot(epochs, metrics["train_loss"], label="training loss")
plt.plot(epochs, metrics["val_loss"], label="validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Baseline training and validation loss")
plt.legend()
plt.show()
```

```python
# Training U-Net

metrics_unet = pd.read_csv('/content/metric_values_unet.csv')
metrics_baseline = pd.read_csv('/content/metric_values_baseline.csv')
epochs = np.array([i for i in range(1, metrics.shape[0]+1)])

plt.figure()
plt.plot(epochs, metrics_unet["mPA"], label="mPA U-Net")
plt.plot(epochs, metrics_unet["mIoU"], label="mIoU U-Net")
plt.plot(epochs, metrics_baseline["mPA"], label="mPA baseline")
plt.plot(epochs, metrics_baseline["mIoU"], label="mIoU baseline")
plt.xlabel("Epochs")
plt.ylabel("Metrics")
plt.title("mPA and mIoU during training")
plt.legend()
plt.show()

plt.figure()
plt.plot(epochs, metrics_unet["train_loss"], label="training loss U-Net")
plt.plot(epochs, metrics_unet["val_loss"], label="validation loss U-Net")
plt.plot(epochs, metrics_baseline["train_loss"], label="training loss baseline")
plt.plot(epochs, metrics_baseline["val_loss"], label="validation loss baseline")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training and validation loss")
plt.legend()
plt.show()
```
